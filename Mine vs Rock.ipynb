# Importing Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Fetching Dataset

```python
data_sonar = pd.read_csv('sonar.all-data.csv', header = None)
```

```python
data_sonar
```

There are 61 columns, out of which ther first 60 columns represents the data attrubutes, and the last column classifies it as Rock or Mine. 

There are 208 rows.

```python
data_sonar[60].value_counts()
```

```python
data_sonar.groupby(60).mean() # Grouping the dataset on the basis of the column 60 on the basis of their mean.
```

```python
X = data_sonar.iloc[ : , : 60] # Feature Vector
Y = data_sonar.iloc[ : , 60 : 61] # Label Space
```

```python
X, Y
```

# Splitting the Dataset following 80 - 20 Rule

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=1) # Following the 80 - 20 Rule
```

```python
X_train
```

```python
X_test
```

# Training Model

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
train_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on training data : ', train_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on testing data : ', test_accuracy)
```
